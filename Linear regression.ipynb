```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.info()
```

```python
cat_df=df.select_dtypes('category')
```

```python
## encoding
```

```python
encoded_cat_df=pd.get_dummies(cat_df,drop_first=True,dtype=int)
```

```python
encoded_cat_df
```

```python
df_1=df.select_dtypes('number')
```

```python
df_1
```

```python
final_df=pd.concat([encoded_cat_df,df_1],axis=1)
```

```python
final_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data=train_test_split(final_df,test_size=0.2,random_state=78)
```

```python
train_data
```

```python
test_data
```

```python
import statsmodels.formula.api as smf
```

```python
final_df.columns
```

```python
'+'.join(final_df.columns)
```

```python
formula='tip~sex_Female+smoker_No+day_Fri+day_Sat+day_Sun+time_Dinner+total_bill+size'
```

```python
model=smf.ols(formula,train_data)
```

```python
model=model.fit()
```

```python
model.summary()
```

```python
model1= smf.ols(formula,train_data).fit()
```

```python
print(model1.summary())
```

```python
## @@ we will check for the signifincance with the p critical and p value and R squared 
```

```python
## we will alway choose the signicant colums / where p value is greater the p critical
```

```python
## since there are only two significant values that size and total_bill
## we will make a new model to improve the model
```

```python
final_df.columns
```

```python
formul2='tip~total_bill+size'
```

```python
model2=smf.ols(formul2,train_data).fit()
```

```python
print(model2.summary())
```

```python
## Here the performace is boosted as R squared has increase 
## R squared is the goodness of fit
```

```python
## always check the signicance with the p value and p critical(0.05)
```

```python
train_predict=model2.predict()
```

```python
test_predict=model2.predict(test_data)
```

```python
train_data['Prediction']=train_predict
```

```python
test_data['Prediction']=test_predict
```
